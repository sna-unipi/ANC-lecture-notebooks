``backboning`` is a python module designed to provide support to network backboning (https://www.michelecoscia.com/?page_id=287).

In this notebook are introduced some of the main features of the module and an overview of its functionalities.

**Note:** this notebook is purposely not 100% comprehensive.

```python
import backboning
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def graph_from_csv(filename):
    g = nx.Graph()
    with open(filename) as f:
        f.readline()
        for l in f:
            l = l.rstrip().split("\t")
            g.add_edge(l[0], l[1], weight=float(l[2]))
    return g
```

```python
def plot_graph(g, pos):
    # Draw the graph with the fixed positions
    plt.figure(figsize=(8, 6))
    nx.draw_networkx(g, pos=pos, with_labels=True, node_color='skyblue', node_size=700, font_size=12, font_color='black', edge_color='gray')
    plt.show()
```

```python
table, nnodes, nnedges = backboning.read("data/karate_club_table.csv", "w", undirected=True, sep="\t")
```

```python
g = graph_from_csv("data/karate_club_table.csv")
pos = nx.spring_layout(g)
plot_graph(g, pos)
```

## Topological Backboning

#### Naive thresholding

```python
naive_table = backboning.naive(table)
naive_backbone = backboning.thresholding(naive_table, 3)
backboning.write(naive_backbone, "backbone", "naive", "data/backbones")
```

```python
g = graph_from_csv("data/backbones/backbone_naive.csv")
plot_graph(g, pos)
```

#### Maximum Spanning Tree

```python
mst_table = backboning.maximum_spanning_tree(table)
backboning.write(mst_table, "backbone", "mst", "data/backbones")
```

```python
g = graph_from_csv("data/backbones/backbone_mst.csv")
plot_graph(g, pos)
```

#### Doubly Stochastic

NB: it has a high computational cost, (implementation) suggested only for small graph.

```python
ds_table = backboning.doubly_stochastic(table)
ds_backbone = backboning.thresholding(ds_table, 0.1)
backboning.write(ds_backbone, "backbone", "ds", "data/backbones")
```

```python
g = graph_from_csv("data/backbones/backbone_ds.csv")
plot_graph(g, pos)
```

#### Doubly High Salience Skeleton

NB: it has a high computational cost, (implementation) suggested only for small graph.

```python
hss_table = backboning.high_salience_skeleton(table)
hss_backbone = backboning.thresholding(hss_table, 0.1)
backboning.write(hss_backbone, "backbone", "hss", "data/backbones")
```

```python
g = graph_from_csv("data/backbones/backbone_hss.csv")
plot_graph(g, pos)
```

## Statistical filtering
#### Noise Corrected

```python
nc_table = backboning.noise_corrected(table, undirected = True, calculate_p_value = True)
nc_backbone = backboning.thresholding(nc_table, 0.005)
backboning.write(nc_backbone, "backbone", "nc", "data/backbones")
```

```python
g = graph_from_csv("data/backbones/backbone_nc.csv")
plot_graph(g, pos)
```

#### Disparity Filter

```python
df_table = backboning.disparity_filter(table)
df_backbone = backboning.thresholding(df_table, 0.005)
backboning.write(df_backbone, "backbone", "df", "data/backbones")
```

```python
g = graph_from_csv("data/backbones/backbone_df.csv")
plot_graph(g, pos)
```
